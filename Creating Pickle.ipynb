```python
import pandas as pd
import glob

path = r'C:\Users\Ashish\Desktop\Capastone Project\hourly_new' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    print(filename)
    df = pd.read_csv(filename, index_col=None, header=0)
    df['Timestamp'] = pd.to_datetime(df['Timestamp']).dt.strftime("%Y%m%d").astype(int)
    df.rename(columns={"Timestamp": "Date"},inplace=True)
    df.set_index('Date')
    k=df.groupby('Date').max() - df.groupby('Date').min() 
    li.append(k)

frame = pd.concat(li, axis=0, ignore_index=False,sort=False)
frame.to_pickle("data_spread")
```
